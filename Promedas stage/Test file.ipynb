```python
import numpy as np 
import matplotlib.pyplot as plt 

N = 1000
d = 1000
x = np.linspace(0,N,N)
y = np.random.choice([-1,0,1],size=[N,d],p=[0.25,0.5,0.25])
z = np.cumsum(y,0)
plt.plot(z,color='lightgrey',linewidth=0.1)
plt.xlim(0,N)
plt.ylim(-1.5*np.sqrt(N),1.5*np.sqrt(N))
plt.plot(x,np.sqrt(x),color='black')
plt.plot(x,-np.sqrt(x),color='black')

```
